```python
import torchvision
import numpy as np
import matplotlib.pyplot as plt
```

```python
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,download=True, transform=None)
```

```python
labels = np.asarray(trainset.targets)
images = np.asarray(trainset.data)
print(labels.shape)
print(images.shape)
mask_cat = np.argwhere(labels==3)
mask_dog = np.argwhere(labels==5)

images_cat = np.squeeze(images[mask_cat])
images_dog = np.squeeze(images[mask_dog])

mask = np.vstack([mask_cat,mask_dog])
print(mask.shape)
print(labels[mask_cat[0]])
print(labels[mask_dog[-1]])
images_cat_dog = images[mask]
print(images_cat_dog.shape)
cat_sample = images_cat[-1]
dog_sample = images_dog[-1]
fig = plt.figure(figsize=[10,5])
plt.subplot(1,2,1)
plt.imshow(cat_sample)
plt.subplot(1,2,2)
plt.imshow(dog_sample)
plt.show()
```
