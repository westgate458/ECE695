```python
import random
import string
```

```python
class People():
    def __init__(self, name_format='first_name_first'):
        
        if name_format not in ['first_name_first', 'last_name_first', 'last_name_with_comma_first']:
            raise NameError('Wrong format')
        
        random.seed(0)
        self.first_names = [''.join([random.choice(string.ascii_lowercase) for _ in range(5)]) for _ in range(10)]
        self.middle_names = [''.join([random.choice(string.ascii_lowercase) for _ in range(5)]) for _ in range(10)]
        self.last_names = [''.join([random.choice(string.ascii_lowercase) for _ in range(5)]) for _ in range(10)]
        
        self.name_format = name_format
        
        self.current = 0
        self.high = 9
        
    def __iter__(self):
        return self

    def __next__(self):
        if self.name_format == 'first_name_first':
            print(self.first_names[self.current]+' '+self.middle_names[self.current]+' '+self.last_names[self.current])
        elif self.name_format == 'last_name_first':
            print(self.last_names[self.current]+' '+self.first_names[self.current]+' '+self.middle_names[self.current])
        else:
            print(self.last_names[self.current]+', '+self.first_names[self.current]+' '+self.middle_names[self.current])
        self.current += 1
        if self.current < self.high:
            return self.current
        raise StopIteration
        
    def __call__(self):
        print(sorted(self.last_names))
    
class PeopleWithMoney(People):
    def __init__(self, name_format='first_name_first'):
        People.__init__(self, name_format)
        self.wealth = [random.randint(0,1000) for _ in range(10)]
        
    def __iter__(self):
        return self
    
    def __next__(self):
        print(self.first_names[self.current]+' '+self.middle_names[self.current]+' '+self.last_names[self.current]+' '+str(self.wealth[self.current]))
        self.current += 1
        if self.current < self.high:
            return self.current
        raise StopIteration
    
    def __call__(self):
        names_full = [[self.wealth[i],self.first_names[i]+' '+self.middle_names[i]+' '+self.last_names[i]] for i in range(10)]
        for ppl in sorted(names_full):
            print(ppl[1],ppl[0])
        
    
```

```python
eg1 = People()
for c in eg1:
    pass
```

```python
eg2 = People('first_name_first')
for c in eg2:
    pass
print('')

eg3 = People('last_name_first')
for c in eg3:
    pass
print('')

eg4 = People('last_name_with_comma_first')
for c in eg4:
    pass
```

```python
eg6 = PeopleWithMoney()
for c in eg6:
    pass
```
